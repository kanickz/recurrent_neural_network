```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

print("TensorFlow version:", tf.__version__)

# Define the vocabulary size
vocab_size = 10000

# Load the IMDB dataset (keeping only the top 10,000 words)
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocab_size)

print(f"Training data sequences: {len(x_train)}")
print(f"Testing data sequences: {len(x_test)}")
```

```python
# Set the maximum length for any review
max_len = 500

# Pad the sequences to ensure identical length
x_train = pad_sequences(x_train, maxlen=max_len)
x_test = pad_sequences(x_test, maxlen=max_len)

print("Shape of x_train after padding:", x_train.shape)
print("Shape of x_test after padding:", x_test.shape)
```

```python
# Create a Sequential RNN model
model = Sequential([
    # Embedding layer: turns word integers into dense vectors of fixed size (128 dimensions)
    Embedding(input_dim=vocab_size, output_dim=128, input_length=max_len),

    # SimpleRNN layer to process the sequential text data
    SimpleRNN(units=64, dropout=0.2, recurrent_dropout=0.2),

    # Dense output layer for binary classification (positive or negative)
    Dense(units=1, activation='sigmoid')
])

# Compile the model with appropriate optimizer, loss, and metrics
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Print out the architecture summary
model.summary()
```

```python
# Train the model
history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.2  # Validating on 20% of the training data
)
```

```python
# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(x_test, y_test)

print("\n--- Final Results ---")
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```
